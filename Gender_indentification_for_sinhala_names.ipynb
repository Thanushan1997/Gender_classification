```python
#impoting pandas package
import pandas as pd
```

```python
#read the male text file and save it in a pandas dataframe
df_males = pd.read_csv('male.txt', delimiter =' ', header = None, names= ['names'])
```

```python
#take the size of the male names
df_males.size
```

```python
#create a list of 1's as the size of the male names 
male_labels = [1]*df_males.size # crea
male_labels
```

```python
#create a column for the label and pass the list of 1's to it, here we label the males names as 1
df_males['label'] = male_labels
```

```python
df_males
```

```python
#read the female text file and save it in a pandas dataframe
df_females = pd.read_csv('female.txt',delimiter = " ", names=['names'])
df_females
```

```python
#create a list of 0's as the size of the female names 
female_labels = [0]*df_females.size
```

```python
#create a column for the label and pass the list of 0's to it, here we label the females names as 0
df_females['label']=female_labels
```

```python
df_females
```

```python
#append both male and female datafam together
Data = df_males.append(df_females)
Data
```

```python
from sklearn.utils import shuffle

# shuffle the dataframe
data = shuffle(Data, random_state=0)
data = data.reset_index(drop=True)#to reset the index of the new dataframe
data.head(20)
```

```python
data.dtypes
```

```python
data.head()
```

```python
#get all the names of to the names feature
names_features  = data['names']
names_features
```

```python
#import the count vectorizer 
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(encoding='utf-8',strip_accents='unicode') #define the vectorizer
X = vectorizer.fit_transform(names_features) #transform the features in numerical form
```

```python
vectorizer.get_feature_names()
```

```python
from sklearn.model_selection import train_test_split
```

```python
Y = data['label']
```

```python
#split the dataset into train and test
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3,random_state=20)
```

```python
#classifiers
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import SGDClassifier
```

```python
from sklearn.metrics import confusion_matrix
```

```python
from sklearn import metrics
import numpy as np

# Define function to calculate the 4 common mertics
def get_metrics(true_labels, predicted_labels):
    
    print('Accuracy:', np.round(
                        metrics.accuracy_score(true_labels, 
                                               predicted_labels),
                        2))
    print('Precision:', np.round(
                        metrics.precision_score(true_labels, 
                                               predicted_labels,
                                               average='weighted'),
                        2))
    print('Recall:', np.round(
                        metrics.recall_score(true_labels, 
                                               predicted_labels,
                                               average='weighted'),
                        2))
    print('F1 Score:', np.round(
                        metrics.f1_score(true_labels, 
                                               predicted_labels,
                                               average='weighted'),
                        2))
```

```python
#define the classifiers
mnb = MultinomialNB()
svm = SGDClassifier(loss='hinge', max_iter=100)
```

```python
#fit the multinomial naive bayes classifer
mnb.fit(X_train,Y_train)
```

```python
#predict the test data using naive bayes classifer
mnb_predictions = mnb.predict(X_test)
```

```python
#call the get metrics function to get the evaluation metrics
get_metrics(true_labels=Y_test,predicted_labels=mnb_predictions)
```

```python
#get the confusion matrix
confusion_matrix(Y_test,mnb_predictions)
```

```python
#train the svm classifer using training data
svm.fit(X_train,Y_train)
```

```python
#predict the test data using Svm
svm_predictions = svm.predict(X_test)
```

```python
#call the get matrix function to get the evaluation matrix
get_metrics(true_labels=Y_test,predicted_labels=svm_predictions)
```

```python
#get the confusion matrix of svm
confusion_matrix(Y_test,svm_predictions)
```

We can see that the accuracy, precision, recall and f1 score are not good for both naive bayes and SVM classifier. Therefore let's try consider last two letters as they best indicators for names in sinhala.

```python
#import the re 
import re
```

```python
#get the list of all the letters in the names using pattern, this is because if we try to get the last two letters directly
#it shows the part of a chracter and that is misleading so we use pattern to get all the letters and then combine them to get the last two letters
names_list = [re.findall(r'\w\W?',name) for name in data['names']]
names_list
```

```python
#define a function to get the last two letters into a list of dictionary 
def lastletters(name):
    list = []
    for i in names_list:
        L_letter = ''.join(i[-1:])
        L2_letter = ''.join(i[-2:])
        list.append({'last_letter' : L_letter,'last2_letters' : L2_letter})
    return list    
```

```python
#now we can get the list of dictionary of last letter and last two letters
data_X = lastletters(names_list)
data_X
```
